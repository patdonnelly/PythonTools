```python
# import necessary databases and libraries
import pycurl,json,requests,sys,os,math
import pandas as pd
import numpy as np
from scipy import stats
# existing redcap API stipulates StringIO, but pycurl now works with BytesIO
from io import BytesIO
```

```python
# get api token
# if you need an access token, consult the RedCap API page
home = os.path.expanduser('~')
with open (home+"/Desktop/redcap_apitoken_repo.txt", "r") as myfile:
    token=myfile.read().split()
```

```python
# say which report you want
# report 13796 is the report in the repository called "Score Report"
rdrp_report = {
    'token': token,
    'content': 'report',
    'format': 'csv',
    'report_id': '13796',
    'rawOrLabel': 'raw',
    'rawOrLabelHeaders': 'raw',
    'exportCheckboxLabel': 'false',
    'returnFormat': 'csv'
}
```

```python
# specify API address
redcap_path = 'https://redcap.iths.org/api/'
# use requests module to pull data from database using api token
# and convert to csv
pull_rdrp = requests.post(redcap_path, data=rdrp_report)
rdrp_report_filename =os.path.join(home+'/Downloads/rdrp.csv')
with open(rdrp_report_filename, 'w') as report_file:
        report_file.write(pull_rdrp.text)
rdrp_data = pd.read_csv(rdrp_report_filename)
```
